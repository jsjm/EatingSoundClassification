### 1. Convert audio data into MEL-spectrogram

```python
# # make a folder for the pngs
# os.mkdir('mel')
# os.mkdir('train')
# os.mkdir('validation')
```

```python
def create_spectrogram(filename, name):
    plt.interactive(False)
    clip, sample_rate = librosa.load(filename, sr=None)
    fig = plt.figure(figsize=[0.72, 0.72]) #param
    ax = fig.add_subplot(111) #param
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_frame_on(False)
    S = librosa.feature.melspectrogram(y=clip, sr=sample_rate)
    librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
    filename = "mel_rd4/" + name + '.png' 
    plt.savefig(filename, dpi=400, bbox_inches='tight', pad_inches=0)
    plt.close()
    fig.clf()
    plt.close(fig)
    plt.close('all')
    del filename, name, clip, sample_rate, fig, ax, S
```

```python
Data_dir=glob("data/clips_cp4/*")

%load_ext memory_profiler

# %%memit 
i=0
for file in Data_dir[i:i+2000]:
    #Define the file.name as is, "name" refers to the png, and is split off into the number itself. 
    filename,name = file,file.split('/')[-1].split('.')[0]
    create_spectrogram(filename,name)
gc.collect()
print ('done 2000')

#%%memit 
i=2000
for file in Data_dir[i:i+2000]:
    filename,name = file,file.split('/')[-1].split('.')[0]
    create_spectrogram(filename,name)
gc.collect()
print ('done 4000')

#%%memit 
i=4000
for file in Data_dir[i:i+2000]:
    filename,name = file,file.split('/')[-1].split('.')[0]
    create_spectrogram(filename,name)
gc.collect()
print ('done 6000')

#%%memit 
i=6000
for file in Data_dir[i:]:
    filename,name = file,file.split('/')[-1].split('.')[0]
    create_spectrogram(filename,name)
gc.collect()
print ('done all')

# #%%memit 
# i=8000
# for file in Data_dir[i:]:
#     filename,name = file,file.split('/')[-1].split('.')[0]
#     create_spectrogram(filename,name)
# gc.collect()
# print ('done all')
```

```python
import soundfile as sf
clips_rd = glob('data/clips_cp4/*')
for clip in clips_rd:
    f = sf.SoundFile(clip)
    t = len(f)/f.samplerate
    if t<4:
        os.remove(clip)
        print('removed: '+ clip)
```
